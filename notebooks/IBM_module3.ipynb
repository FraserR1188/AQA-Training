```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
%matplotlib inline
```

```python
filepath="https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-Coursera/laptop_pricing_dataset_mod2.csv"
df = pd.read_csv(filepath, header=0)
```

```python
df.head(10)
```

```python
"""Task 1 - Visualize individual feature patterns
Continuous valued features
Generate regression plots for each of the parameters "CPU_frequency", "Screen_Size_inch" and 
"Weight_pounds" against "Price". Also, print the value of correlation of each feature with "Price"."""
```

```python
# CPU_frequency plot
sns.regplot(x='CPU_frequency', y='Price', data=df)
plt.ylim(0,)
```

```python
# Screen_Size_inch plot
sns.regplot(x='Screen_Size_inch', y='Price', data=df)
plt.ylim(0,)
```

```python
# Weight_pounds plot
sns.regplot(x='Weight_pounds', y='Price', data=df)
plt.ylim(0,)
```

```python
# Correlation values of the three attributes with Price

for param in ['CPU_frequency', 'Screen_Size_inch', 'Weight_pounds']:
    print(f'Correlation of Price and {param} is ', df[[param, 'Price']].corr())
```

```python
"""Categorical features
Generate Box plots for the different feature that hold categorical values. 
These features would be "Category", "GPU", "OS", "CPU_core", "RAM_GB", "Storage_GB_SSD"."""
```

```python
# Category Box plot
sns.boxplot(x='Category', y='Price', data=df)
```

```python
# GPU Box plot
sns.boxplot(x='GPU', y='Price', data=df)
```

```python
# OS Box plot
sns.boxplot(x='OS', y='Price', data=df)
```

```python
# CPU_core Box plot
sns.boxplot(x='CPU_core', y='Price', data=df)
```
